```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

## 서울시 구별 가로수 / 공원율 / 도로율 시각화 (2016 - 2020)

```python
dust = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/dustdata/seoul_dust.csv')
dust 
```

```python
cols = dust.columns[1:]
cols
```

```python
plt.figure(figsize=(24,12))
for i, col in enumerate(cols):

    plt.subplot(3,2,i+1)
    plt.xticks(rotation=45)
    sns.barplot(dust['gu'], dust[col])

plt.show()
```

## 서울시 구별 가로수 / 공원율 / 도로율 시각화 (2020)

```python
dust20 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/dustdata/seoul_dust20.csv')
```

```python
cols = dust20.columns[1:]

plt.figure(figsize=(24,12))
for i, col in enumerate(cols):

    plt.subplot(3,2,i+1)
    plt.xticks(rotation=45)
    sns.barplot(dust20['gu'], dust20[col])

plt.show()
```

### Plotly

```python
import plotly.express as px
```

```python

fig = px.bar(dust, x='gu', y='count',color='gu', text='count' )
fig.show()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
cols = dust.columns[1:]

fig = make_subplots(rows=3, cols=2, 
                    subplot_titles=(cols[0],cols[1],cols[2],cols[3],cols[4]) )
for i, col in enumerate(cols):
    c = 2 if i%2 else 1
    r = i // 2 + 1   
    fig.add_trace(go.Bar(x=dust['gu'], y=dust[col]),
                row=r, col=c)
    fig.update_xaxes(tickangle=45)
    fig.update_layout(width=1100, height=700,)
    
fig.show()
```

```python
cols = dust20.columns[1:]

fig = make_subplots(rows=3, cols=2,
                    subplot_titles=(cols[0],cols[1],cols[2],cols[3],cols[4]) )
for i, col in enumerate(cols):
    c = 2 if i%2 else 1
    r = i // 2 + 1   
    fig.add_trace(go.Bar(x=dust20['gu'], y=dust20[col]),
                row=r, col=c)
    fig.update_xaxes(tickangle=45)
    fig.update_layout(width=1100, height=700)
    
fig.show()
```
