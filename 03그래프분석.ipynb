## 서울시 구별 가로수 / 공원율 / 도로율 시각화 (2016 - 2020)

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
cols = dust.columns[1:]

fig = make_subplots(rows=3, cols=2, 
                    subplot_titles=('미세먼지 나쁨일수',cols[1],cols[2],cols[3],cols[4]) )
for i, col in enumerate(cols):
    c = 2 if i%2 else 1
    r = i // 2 + 1   
    fig.add_trace(go.Bar(x=dust['gu'], y=dust[col]),
                row=r, col=c)
    fig.update_xaxes(tickangle=45)
    fig.update_layout(width=1100, height=700,)
    
fig.show()
```

## 상관 관계 분석

```python
dust = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/dustdata/seoul_dust1620.csv')
dust.rename(columns={'gu': '자치구', 'count':'미세먼지나쁨일수'},inplace=True)
dust 
```

```python
plt.scatter(dust['공원율'], dust['미세먼지나쁨일수'])
plt.xticks(rotation=90)
plt.show()
```

```python
dust.drop('연도', inplace=True, axis=1)
dust.head()
```

```python

_, axs = plt.subplots(2, 2, figsize=(18,8))
for i, feature in enumerate(dust.columns[2:]):
    row, col = i //2, i % 2
    sns.regplot(x=feature, y='미세먼지나쁨일수', data=dust, ci=None, ax=axs[row][col])
```
