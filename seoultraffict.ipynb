```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
seoult = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/seoul_traffic1620.csv')
seoult.tail()
```

```python
st = seoult.pivot('자치구', '연도', '교통량')
st.head()
```

```python
# heatmap by plt.pcolor()
plt.pcolor(st)
plt.xticks(np.arange(0.5, len(st.columns), 1), st.columns)
plt.yticks(np.arange(0.5, len(st.index), 1), st.index)
plt.title('서울시 교통량', fontsize=20)
plt.xlabel('자치구', fontsize=14)
plt.ylabel('연도', fontsize=14)
plt.colorbar()

st.style.background_gradient(cmap='summer')

plt.show()
```
